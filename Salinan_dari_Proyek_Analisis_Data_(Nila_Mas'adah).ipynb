Untuk memastikan kualitas data, akan dilakukan pemeriksaan data untuk mengidentifikasi masalah yang terdapat dalam DataFrame. Pemeriksaan tersebut dilakukan dengan:

#### ✅ **Cleaning Data**

Berdasarkan hasil proses assesing data, diketahui bahwa terdapat 3 masalah berbeda pada DataFrame, yaitu tipe data, missing value, dan duplicate data.
- Untuk mengatasi masalah tipe data, akan dilakukan penggantian tipe data pada kolom yang memiliki kesalahan tipe data. Pertama, dibuat sebuah list yang berisi nama kolom yang ingin diubah tipe datanya dan disimpan dalam variabel baru. Selanjutnya, digunakan loop `for`, setiap kolom dalam list baru akan dikonversi menjadi tipe data yang sesuai dengan fungsi tipe data tertentu. Terakhir, periksa menggunakan method `info()` untuk memastikan bahwa tipe data lama telah berubah menjadi tipe data yang baru .
- Untuk mengatasi missing value, pertama akan ditampilkan semua baris dari DataFrame yang memiliki nilai `NaN` pada kolom tertentu dengan method `isna()`. Apabila proporsi jumlah missing value sangat kecil dibandingkan dari jumlah total entri, maka menggunakan method `dropna()` untuk menghapus seluruh baris tidak akan berdampak signifikan. Parameter `axis=0` menandakan keinginan men-drop seluruh baris yang mengandung missing value. Namun, apabila proporsi jumlah missing value sangat besar dibandingkan dari jumlah total entri, maka digunakan method `fillna()` untuk mengganti missing value dengan nilai tertentu. `inplace=True` menandakan keinginan langsung menerapkan operasi tersebut dalam DataFrame.
- Untuk menghapus duplicate data, akan digunakan method `drop.duplicates()` dan memastikan kembali jumlah duplicate data setelah dihapus menggunakan method `duplicated()` dan `sum()`.

## ▶ **Exploratory Data Analysis (EDA)**

Tahap ini bertujuan untuk memperoleh insight dari DataFrame yang telah dibersihkan sehingga dapat menjawab pertanyaan bisnis yang diajukan. Langkah pertama adalah mengeksplorasi parameter statistik dari DataFrame menggunakan method `describe()` dan parameter `include="all"`. Method dan parameter ini dapat digunakan untuk melihat rangkuman parameter statistik, seperti count, unique, mean, max, min, dan standar deviasi. Langkah selanjutnya adalah menggabungkan data dengan teknik `relational database` dimana memanfaatkan primary key dan foreign key selama proses merging. Merging data dilakukan karena terdapat informasi bisnis yang tersebar di berbagai tabel dan ditemukannya relasi di lebih dari satu tabel untuk menjawab pertanyaan bisnis. Dengan menggunakan fungsi `merge()` berjenis `inner` membuat proses merging hanya akan mengambil nilai yang bersesuaian di dua atau lebih tabel. Setelah merging data selesai, akhirnya dapat dilakukan pengelompokkan data. Pengelompokkan data ini menggunakan konsep pivot tabel untuk menghitung parameter statistik dari kelompok data tertentu. Selama pembuatan pivot tabel dibantu method `groupby()` dan `agg()`.

## ▶ **Analisis Lanjutan**

Dalam analisis lanjutan akan diterapkan teknik RFM Analysis. RFM analysis merupakan salah satu metode untuk melakukan segmentasi pelanggan (mengelompokkan pelanggan ke dalam beberapa kategori) berdasarkan 3 parameter, yaitu recency, frequency, dan monetary.
- Recency: parameter untuk melihat kapan terakhir seorang customer melakukan transaksi. Semakin kecil nilai recency, semakin baru transaksi terakhirnya, yang menunjukkan bahwa customer masih aktif.
- Frequency: parameter untuk mengidentifikasi seberapa sering seorang customer melakukan transaksi. Semakin tinggi nilai frequency, semakin sering customer berinteraksi dengan bisnis, yang menunjukkan loyalitasnya.
- Monetary: parameter untuk mengidentifikasi seberapa besar revenue yang berasal dari customer tersebut. Semakin tinggi nilai monetary, semakin besar kontribusi finansial customer tersebut terhadap bisnis.

#### **Pertanyaan Bisnis RFM Analysis**

2. Seberapa sering customer melakukan order dalam beberapa bulan terakhir?
3. Berapa banyak pengeluaran yang dihabiskan customer dalam beberapa bulan terakhir?
